```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(1)
ax = fig.add_subplot(111)

# N=2
xx = np.arange(0,1,0.01)
ax.plot(xx, xx/4, label='N=2')

# N=3
xx = np.array((0.01,0.1,0.3,0.5,0.7,1.0))
yy = np.array((1.069,1.072,1.075,1.077,1.068,1.072))
yerr = 0.002
#ax.errorbar(xx, yy*xx, yerr=yerr, fmt='o', label='N=3')
ax.plot(xx, yy*xx, label='N=3')

# N=4
yy = np.array((2.797, 2.799, 2.793, 2.783, 2.787, 2.791))
ax.plot(xx, yy*xx, label='N=4')

# N=5
yy = np.array((5.764, 5.741, 5.73, 5.71, 5.762, 5.72))
ax.plot(xx, yy*xx, label='N=5')

# N=6
yy = np.array((10.21, 10.12, 10.2, 10.14, 10.24, 10.19))
ax.plot(xx, yy*xx, label='N=6')

ax.set_xlabel(r'$B_N^2$')
ax.set_ylabel(r'$|E_\mathrm{min}/m_Q|$')
ax.legend()
plt.savefig('EminVB2.pdf',transparent='true')
plt.show()
```

# Defining SUNc color factors:

```python
import sympy as sp


def CA(Nc):
    return Nc

def CF(Nc):
    return (Nc**2 - 1) / (2 * Nc)

def dAdA(Nc,NA):
    return Nc**2 * (Nc**2 + 36) * NA / 24

def dFdA(Nc,NA):
    return Nc * (Nc**2 + 6) * NA / 48

def dFdF(Nc,NA):
    return (Nc**4 - 6 * Nc**2 + 18) * NA / (96 * Nc**2)

def dAA(Nc):
    return dAdA(Nc,1)

def dFA(Nc):
    return dFdA(Nc,1)

def dFF(Nc):
    return dFdF(Nc,1)

Tf = 1/2
```

```python
import numpy as np
import scipy.special as spl

Pi = np.pi


#The 2 first/scheme-independent

def Beta0(Nf, Nc=3):
    Beta0 = (1 / (4 * Pi)) * (11/3 * CA(Nc) - 2/3 * Nf)
    return Beta0

def Beta1(Nf, Nc=3):
    Beta1 = (1 / (4 * Pi))**2 * (34/3 * CA(Nc)**2 - 10/3 * CA(Nc) * Nf - 2 * CF(Nc) * Nf)
    return Beta1
```

```python
def LambdaQCD_MSbar(f):
    if f == 3:
        return 0.338
    elif f == 4:
        return 0.296
    elif f == 5:
        return 0.213
    elif f == 6:
        return 0.090
    else:
        return None  # Undefined for other values

def LQ_MSbar(Q, f):
    return np.log(Q**2 / LambdaQCD_MSbar(f)**2)
```

```python
def MAlpha_s(Q, f):
    alpha_s = Alpha_sNf(Q, f)
    return 1 + (-0.291667) * (alpha_s / math.pi)**2 + (-5.32389 + (f - 1) * 0.26247) * (alpha_s / math.pi)**3

def MAlpha_s2Loop(Q, f):
    return 1 + (-0.291667) * (Alpha_sNf(Q, f) / math.pi)**2
```

```python
def Alpha_s(Q):
    if Q < 1:
        return MAlpha_s(Mc, 4) * MAlpha_s(Mb, 5) * Alpha_sNf(1, 5)
    elif 1 <= Q < Mc:
        return MAlpha_s(Mc, 4) * MAlpha_s(Mb, 5) * Alpha_sNf(Q, 5)
    elif Mc <= Q < Mb:
        return MAlpha_s(Mb, 5) * Alpha_sNf(Q, 5)
    elif Mb <= Q < Mt:
        return Alpha_sNf(Q, 5)
    elif Q >= Mt:
        return 1 / MAlpha_s(Mt, 6) * Alpha_sNf(Q, 5)
    else:
        return None

def Alpha_s3Loop(Q):
    if Q < 1:
        return MAlpha_s2Loop(Mc, 4) * MAlpha_s2Loop(Mb, 5) * Alpha_sNf3Loop(1, 5)
    elif 1 <= Q < Mc:
        return MAlpha_s2Loop(Mc, 4) * MAlpha_s2Loop(Mb, 5) * Alpha_sNf3Loop(Q, 5)
    elif Mc <= Q < Mb:
        return MAlpha_s2Loop(Mb, 5) * Alpha_sNf3Loop(Q, 5)
    elif Mb <= Q < Mt:
        return Alpha_sNf3Loop(Q, 5)
    elif Q >= Mt:
        return 1 / MAlpha_s2Loop(Mt, 6) * Alpha_sNf3Loop(Q, 5)
    else:
        return None

def Alpha_s2Loop(Q):
    if Q < 1:
        return Alpha_sNf2Loop(1, 3)
    elif 1 <= Q < Mc:
        return Alpha_sNf2Loop(Q, 3)
    elif Mc <= Q < Mb:
        return Alpha_sNf2Loop(Q, 4)
    elif Mb <= Q < Mt:
        return Alpha_sNf2Loop(Q, 5)
    elif Q >= Mt:
        return Alpha_sNf2Loop(Q, 6)
    else:
        return None

def Alpha_s1Loop(Q):
    if Q < 1:
        return Alpha_sNf1Loop(1, 3)
    elif 1 <= Q < Mc:
        return Alpha_sNf1Loop(Q, 3)
    elif Mc <= Q < Mb:
        return Alpha_sNf1Loop(Q, 4)
    elif Mb <= Q < Mt:
        return Alpha_sNf1Loop(Q, 5)
    elif Q >= Mt:
        return Alpha_sNf1Loop(Q, 6)
    else:
        return None
```

```python
def LambdaQCD_MSbar(f):
    if f == 3:
        return 0.338
    elif f == 4:
        return 0.296
    elif f == 5:
        return 0.213
    elif f == 6:
        return 0.090
    else:
        return None  # Undefined for other values

def LR_MSbar(R, f):
    return np.log(1 / (R**2 * LambdaQCD_MSbar(f)**2))

###########################################################
def Alpha_sNf(R, f):
    LR_val = LR_MSbar(R, f)
    #print('oooooooooooooo',LR_MSbar(Mb, f))
    #print('iuhgbjkuh',np.log(LR_val))
    #print('oiuhghgggggg',np.log(LR_MSbar(Mb, f)))
    return 1 / (Beta0(3,f) * LR_val) - \
           Beta1(3,f) * np.log(LR_val) / (Beta0(3,f)**3 * LR_val**2) + \
           1 / (Beta0(3,f)**3 * LR_val**3) * (Beta1(3,f)**2 / Beta0(3,f)**2 * (np.log(LR_val)**2 - np.log(LR_val) - 1) + Beta2(3,f) / Beta0(3,f)) + \
           1 / (Beta0(3,f)**4 * LR_val**4) * (Beta1(3,f)**3 / Beta0(3,f)**3 * (-np.log(LR_val)**3 + 2.5 * np.log(LR_val)**2 + 2 * np.log(LR_val) - 0.5) - \
           3 * Beta1(3,f) * Beta2(3,f) * np.log(LR_val) / Beta0(3,f)**2 + Beta3(3,f) / (2 * Beta0(3,f)))


def Alpha_sNf3Loop(R, f):
    LR_val = LR_MSbar(R, f)
    return 1 / (Beta0(3,f) * LR_val) - \
           Beta1(3,f) * np.log(LR_val) / (Beta0(3,f)**3 * LR_val**2) + \
           1 / (Beta0(3,f)**3 * LR_val**3) * (Beta1(3,f)**2 / Beta0(3,f)**2 * (np.log(LR_val)**2 - np.log(LR_val) - 1) + Beta2(3,f) / Beta0(3,f))

def Alpha_sNf2Loop(R, f):
    LR_val = LR_MSbar(R, f)
    return 1 / (Beta0(3,f) * LR_val) - \
           Beta1(3,f) * np.log(LR_val) / (Beta0(3,f)**3 * LR_val**2)

def Alpha_sNf1Loop(R, f):
    LR_val = LR_MSbar(R, f)
    return 1 / (Beta0(3,f) * LR_val)


def MAlpha_s(Q, f):
    #print('ok')
    alpha_s = Alpha_sNf(1/Q, f)
    return 1 + (-0.291667) * (alpha_s / math.pi)**2 + (-5.32389 + (f - 1) * 0.26247) * (alpha_s / math.pi)**3

def MAlpha_s2Loop(Q, f):
    return 1 + (-0.291667) * (Alpha_sNf(1/Q, f) / math.pi)**2


##############################################################

def Alpha_s(R):
    if 1/R < 1:
        return MAlpha_s(Mc, 4) * MAlpha_s(Mb, 5) * Alpha_sNf(1, 3)
    elif 1 <= 1/R < Mc:
        return MAlpha_s(Mc, 4) * MAlpha_s(Mb, 5) * Alpha_sNf(R, 3)
    elif Mc <= 1/R < Mb:
        return MAlpha_s(Mb, 5) * Alpha_sNf(R, 4)
    elif Mb <= 1/R < Mt:
        return Alpha_sNf(R, 5)
    elif 1/R >= Mt:
        return 1 / MAlpha_s(Mt, 6) * Alpha_sNf(R, 6)
    else:
        return None

def Alpha_s1Loop(R):
    if 1/R < 1:
        return Alpha_sNf1Loop(1, 3)
    elif 1 <= 1/R < Mc:
        return Alpha_sNf1Loop(R, 3)
    elif Mc <= 1/R < Mb:
        return Alpha_sNf1Loop(R, 4)
    elif Mb <= 1/R < Mt:
        return Alpha_sNf1Loop(R, 5)
    elif 1/R >= Mt:
        return Alpha_sNf1Loop(R, 6)
    else:
        return None

def Alpha_s2Loop(R):
    if 1/R < 1:
        return Alpha_sNf2Loop(1, 3)
    elif 1 <= 1/R < Mc:
        return Alpha_sNf2Loop(R, 3)
    elif Mc <= 1/R < Mb:
        return Alpha_sNf2Loop(R, 4)
    elif Mb <= 1/R < Mt:
        return Alpha_sNf2Loop(R, 5)
    elif 1/R >= Mt:
        return Alpha_sNf2Loop(R, 6)
    else:
        return None

def Alpha_s3Loop(R):
    if 1/R < 1:
        return MAlpha_s2Loop(Mc, 4) * MAlpha_s2Loop(Mb, 5) * Alpha_sNf3Loop(1, 3)
    elif 1 <= 1/R < Mc:
        return MAlpha_s2Loop(Mc, 4) * MAlpha_s2Loop(Mb, 5) * Alpha_sNf3Loop(R, 3)
    elif Mc <= 1/R < Mb:
        return MAlpha_s2Loop(Mb, 5) * Alpha_sNf3Loop(R, 4)
    elif Mb <= 1/R < Mt:
        return Alpha_sNf3Loop(R, 5)
    elif 1/R >= Mt:
        return 1 / MAlpha_s2Loop(Mt, 6) * Alpha_sNf3Loop(R, 6)
    else:
        return None
```
